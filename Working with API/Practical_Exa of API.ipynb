```python
import requests as req
base_url = "https://pokeapi.co/api/v2/"
def get_info(name):
    url = f'{base_url}/pokemon/{name}'
    response = req.get(url)
    if response.status_code == 200:
        charater_data = response.json()
        return charater_data
    else:
        print("Failed to Fetch Data.")

charter_name = 'jigglypuff'
charter_info = get_info(charter_name)
if charter_info != None:
    print(f"Name: {charter_info['name']}")
    print(f"Height: {charter_info['height']}")
    print(f"Weight: {charter_info['weight']}")
```
