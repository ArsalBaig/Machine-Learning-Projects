```python
import requests

base_URL = 'https://pokeapi.co/api/v2/'
def get_info(name):
    url = f'{base_URL}/pokemon/{name}'
    response = requests.get(url)
    if response.status_code == 200:
      print('Data Retrived Successfully.')
    else:
        print('Failed to Retrive Data.{response.status_code}')

poki_name = 'pikachu'
get_info(poki_name)
```

```python
from os import name
import requests

base_URL = 'https://pokeapi.co/api/v2/'
def get_info(name):
    url = f'{base_URL}/pokemon/{name}'
    response = requests.get(url)
    if response.status_code == 200:
      pokemon_data = response.json()
      return pokemon_data
    else:
        print('Failed to Retrive Data.{response.status_code}')

poki_name = 'pikachu'
pokemon_data = get_info(poki_name)
if pokemon_data:
    print(f"Name: {pokemon_data['name']}")
    print(f"Weight: {pokemon_data['weight']}")
    print(f"Height: {pokemon_data['height']}")
```
