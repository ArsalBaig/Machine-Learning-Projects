```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
web_Url = requests.get('https://www.pakwheels.com/used-cars/imported/57428', headers= headers).text
```

```python
soup = BeautifulSoup(web_Url, 'lxml')
```

```python
cars_details = soup.find_all('div', class_ = 'well search-list clearfix ad-container page-')
```

```python
len(cars_details)
```

```python
model = []
variant = []
power = []
price = []
for i in cars_details:
    model.append(i.find('h3').text.strip())
    variant.append(i.find_all('ul', class_ = 'list-unstyled search-vehicle-info-2 fs13')[0].text.strip())
    price.append(i.find('div', class_ = 'price-details generic-dark-grey').text.strip())
    power.append(i.find_all('ul', class_ = 'list-unstyled search-vehicle-info-2 fs13')[0].text.strip())

d = {'Model': model, 'Variant': variant, 'Price': price}
df = pd.DataFrame(d)
```

```python
# Creating Automation for next pages searching.
final = pd.DataFrame()

for j in range(1, 15):
    web_Url = f'https://www.pakwheels.com/used-cars/imported/57428/page/{j}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(web_Url, headers= headers).text
    soup = BeautifulSoup(web_Url, 'lxml')
    cars_details = soup.find_all('div', class_ = 'well search-list clearfix ad-container page-')

model = []
variant = []
power = []
price = []
for i in cars_details:
    model.append(i.find('h3').text.strip())
    variant.append(i.find_all('ul', class_ = 'list-unstyled search-vehicle-info-2 fs13')[0].text.strip())
    price.append(i.find('div', class_ = 'price-details generic-dark-grey').text.strip())
    power.append(i.find_all('ul', class_ = 'list-unstyled search-vehicle-info-2 fs13')[0].text.strip())

d = {'Model': model, 'Variant': variant, 'Price': price}
df = pd.DataFrame(d)

final = pd.concat([final, df], ignore_index = True)
```
